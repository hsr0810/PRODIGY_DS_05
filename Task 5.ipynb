```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X=pd.read_csv("data.csv")
```

```python
# MISSING_VALUES

missing_values = X.isnull().sum()
print("Missing Values in Each Column:\n", missing_values)
```

```python
# DUBLICATE_VALUES

duplicates = X.duplicated().sum()
print("Duplicate Entries: ", duplicates)
```

```python
# Visualize accident distribution by Weather conditions

plt.figure(figsize=(16,8))
sns.countplot(x='Weather_Condition', data=X)
plt.xticks(rotation=90)
plt.xlabel('Weather Type')
plt.ylabel('Number of Accidents')
plt.title('Accidents by Weather Type')
plt.show()
```

```python
# Visualize accident distribution by Traffic Signal

plt.figure(figsize=(12,6))
sns.countplot(x='Traffic_Signal', data=X)
plt.xlabel('Traffic_Signal')
plt.ylabel('Number of Accidents')
plt.title('Accidents by Traffic_Signal')
plt.show()
```

```python
# Visualize accident distribution by Rain

plt.figure(figsize=(12,6))
sns.countplot(x='Rain', data=X)
plt.xlabel('Traffic_Signal')
plt.ylabel('Number of Accidents')
plt.title('Accidents on Rain Weather')
plt.show()
```

```python
# Visualize accident distribution by Cloud

plt.figure(figsize=(12,6))
sns.countplot(x='Cloud', data=X)
plt.xlabel('Traffic_Signal')
plt.ylabel('Number of Accidents')
plt.title('Accidents on Cloud Weather')
plt.show()
```

```python
# Visualize accident distribution by Clear Weather

plt.figure(figsize=(12,6))
sns.countplot(x='Clear', data=X)
plt.xlabel('Traffic_Signal')
plt.ylabel('Number of Accidents')
plt.title('Accidents on Clear Weather')
plt.show()
```

```python
# Visualize accident distribution on Heavy_Rain

plt.figure(figsize=(12,6))
sns.countplot(x='Heavy_Rain', data=X)
plt.xlabel('Traffic_Signal')
plt.ylabel('Number of Accidents')
plt.title('Accidents on Heavy_Rain Weather')
plt.show()
```
